```python
# load functions
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Dark2_6

# load data
dfc =pd.read_csv('~/Documents/Neurohacker/EEG/Batch2/clean-data/museMonitor_2018-03-17--08-05-18.csv')
dfr =pd.read_csv('~/Documents/Neurohacker/EEG/Batch2/raw-data/museMonitor_2018-03-17--08-05-18.csv')
dfr =pd.read_csv('~/Documents/Neurohacker/EEG/Batch2/raw-data/museMonitor_2018-03-17--08-41-30.csv')

# remove nan rows
dfr=dfr[np.isfinite(dfr['Delta_TP9'])]
# check        
dfr.head()
```

```python
# convert timestamps to seconds
# default timestamps as date format
dfr.dtypes
dfr_t=dfr['TimeStamp']

# split date and time
dfr_t=dfr_t.str.split(' ',expand=True)
dft_tt=dfr_t[1].str.split(':',expand=True)
# convert object to numeric format
dft_tt[0]=pd.to_numeric(dft_tt[0])
dft_tt[1]=pd.to_numeric(dft_tt[1])
dft_tt[2]=pd.to_numeric(dft_tt[2])
# convert to s
dft_tt=dft_tt[0]*60*60+dft_tt[1]*60+dft_tt[2]
# check
#dft_tt.head()
```

```python
#check
#dfr.columns
```

```python
# Generate data for plotting

# choose electrode to plot
# replace with name of electrode
chooseelect='AF7'

# generate new plot dataframe
# timestamps
xval=dft_tt; 
# data for different columns
yvalselect=[col for col in dfr.columns if chooseelect in col]
yval=dfr[yvalselect]
yval=yval.iloc[:,:4]
#yval.head()

# get EEG band names
yvalcolname=[col for col in yval.columns]
yvalcolname=pd.Series(yvalcolname)
#[yvalcolname.split('_',expand=True)]
yvalcolname=yvalcolname.str.split('_',expand=True)
#yvalcolname
```

```python
# plot
output_notebook()
p = figure(plot_width=800, plot_height=400,title=print("Electrode: {}".format(chooseelect)))

# specify color of each line
nlines=len(yval.columns)
ncolor=Dark2_6[0:nlines]

for x in range(0, nlines):
    p.line(xval,yval.iloc[:,x].values ,line_color=ncolor[x],legend=yvalcolname.iloc[x,0])
    
p.axis.axis_label='Seconds'
p.yaxis.axis_label='Power (units uknown)'
show(p)
```
