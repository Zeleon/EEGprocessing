```python
# ####load functions

import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Dark2_6
```

```python
# ####load data

# ---Replace with data path---
loadpath='~/Documents/Neurohacker/EEG/data/Batch2/'

# ---Replace with file name---
loadfile='museMonitor_2018-03-17--08-05-18.csv'
loadfile='museMonitor_2018-03-17--08-19-49.csv'
#loadfile='museMonitor_2018-03-17--08-41-30.csv'

# load
loadpathfile_clean="{}clean-data/{}".format(loadpath,loadfile)
loadpathfile_raw="{}raw-data/{}".format(loadpath,loadfile)
dfc =pd.read_csv(loadpathfile_clean)
dfr =pd.read_csv(loadpathfile_raw)

dfr.head()
```

```python
# ####Select data

chooseelectlist=['AF7','AF8','TP9','TP10']
# ---replace with choice---
chooseelect=0
chooseelect=chooseelectlist[chooseelect]

# get columns with chosen electrode
yvalselect=[col for col in dfr.columns if chooseelect in col]

# remove nan rows
dfr=dfr[np.isfinite(dfr[yvalselect[1]])]
# check        
yval=dfr[yvalselect]
yval.head()
```

```python
# ####convert timestamps to seconds

# default timestamps as date format
dfr.dtypes
dfr_t=dfr['TimeStamp']

# split date and time
dfr_t=dfr_t.str.split(' ',expand=True)
dft_tt=dfr_t[1].str.split(':',expand=True)
# convert object to numeric format
dft_tt[0]=pd.to_numeric(dft_tt[0])
dft_tt[1]=pd.to_numeric(dft_tt[1])
dft_tt[2]=pd.to_numeric(dft_tt[2])
# convert to s
dft_tt=dft_tt[0]*60*60+dft_tt[1]*60+dft_tt[2]
# check
#dft_tt.head()
```

```python
# ####Generate data for plotting

# generate new plot dataframe
# timestamps
xval=dft_tt; 

# which EEG band to plot
EEGbandtoplot=['Delta','Theta','Alpha','Beta']

# get EEG band names
yvalcolname=[col for col in yval.columns]
yvalcolname=np.array(yvalcolname)
# find which columns has substring from EEGbandtoplot
yvalcolname=yvalcolname[[np.array([any([y in x for y in EEGbandtoplot]) for x in yvalcolname])]].tolist()

# select correct columns in data
yval=yval[yvalcolname]

# generate legend
yvalcolname=[x.split('_') for x in yvalcolname]
```

```python
# ####plot
output_notebook()
p = figure(plot_width=800, plot_height=400,title=print("Electrode: {}".format(chooseelect)))

# specify color of each line
nlines=len(yval.columns)
ncolor=Dark2_6[0:nlines]

for x in range(0, nlines):
    p.line(xval,yval.iloc[:,x].values ,line_color=ncolor[x],legend=yvalcolname[x][0])
    
p.axis.axis_label='Seconds'
p.yaxis.axis_label='Power (units unknown)'
show(p)
```
